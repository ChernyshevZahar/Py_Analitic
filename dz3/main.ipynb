## Вывести на экран первые 5 строк

```python
data = pd.read_csv("kc-house-data.csv")
data.head(5)
```

 ### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома Если реновации дома не было, то в новом признаке поставьте 0

```python
data['delta_renovated'] = data['yr_renovated'] - data['yr_built']  
data['delta_renovated'] = data['delta_renovated'].apply(lambda x: max(x, 0))
data.head(5)
```

### 1.3 Создайте признаки года продажи, месяца продажи

```python
data['year_sold'] = pd.to_datetime(data['date']).dt.year  
data['month_sold'] = pd.to_datetime(data['date']).dt.month 
data.head(5)
```

### 1.4 Удалите признаки date, zipcode, lat, long


```python
data = data.drop(['date', 'zipcode', 'lat', 'long'], axis=1)
data.head(5)
```

## 3 задача
Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
avg_price_by_bedrooms = data.pivot_table(index='bedrooms', values='price', aggfunc='mean').sort_values(by='price', ascending=True)  
avg_price_by_bedrooms.head()
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
stats_by_condition = data.pivot_table(index='condition', values='price', aggfunc=['min', 'mean', 'max'])  
stats_by_condition.head()
```

### 3.3 Постройте таблицу с подсчетом количества домов в зависимости от вида на набережную и оценки вида:

```python
count_by_water_view = data.pivot_table(index='waterfront', columns='grade', values='price', aggfunc='count', fill_value=0)  
count_by_water_view.head()
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
count_by_floors_bedrooms = data.pivot_table(index='floors', columns='bedrooms', values='price', aggfunc='count', fill_value=0)  
count_by_floors_bedrooms.head() 
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в зависимости от состояния дома и оценки дома:

```python
median_price_by_condition_grade = data.pivot_table(index='condition', columns='grade', values='price', aggfunc='median')  
median_price_by_condition_grade.head()
```
